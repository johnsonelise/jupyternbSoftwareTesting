##### Read a Jupyter Notebook File

```python
import nbformat
import sys
import matplotlib.pyplot as plt
import random
import glob
import os.path
from nbformat.reader import NotJSONError
from json.decoder import JSONDecodeError

folder_path = "./went_through_mine_worked/"

# Create a pattern to match all files in the folder
pattern = os.path.join(folder_path, '*')
count = 0
# Specify the path to the Jupyter Notebook file
# notebook_file_path = 'outs/order-testing.ipynb'
original_stdout = sys.stdout
# sys.stdout = open('testing_only_valid.txt', 'w')
print('starting...')
# print(f'Number of Files: {len(pattern)}')
plt.switch_backend('Agg')
file_number = 0
total_num_cells = []
lines_changed = []

lines_nb = []

def setup(notebook_file_path):
    try:
        with open(notebook_file_path, 'r', encoding='utf-8') as notebook_file:
            notebook_content = nbformat.read(notebook_file, as_version=4)
    except (JSONDecodeError, NotJSONError) as e:
        print(f"Error reading notebook at {notebook_file_path}: {e}")
        return None  # Return None to indicate an error
    
    code = []
    num_cells  = 0
    count = 0
    # read everything in the file!
    for cell in notebook_content['cells']:
        if cell.cell_type == 'code':
            code_source = cell['source']
            num_cells += 1
            
            i = 1
            og = str(code_source).count('\n')
            while str(code_source).count('\n') < 10 and i < 12:
                if len(code) >= 0 + i:
                    code[len(code) - i] += (str(code_source) + '\n')
                    count = str(code_source).count('\n')
                    print(f'changed -- elise: {count}')
                    i += 1
                else: 
                    i = 15

            
            code.append(str(code_source) + '\n')

            lines_changed.append(abs(og - count))
        elif cell.cell_type == 'markdown':
            markdown_source = cell['source']

    print(f'num cells: {num_cells}') 
    total_num_cells.append(num_cells)
    return code
```

#### Run code for all files in the folder!

```python
# Iterate through the files

for file_path in glob.glob(pattern):
    print(count)
    file_number = file_number + 1
    print(f'file number: {file_number}')
    print(file_path[27:])
    # Check if it's a file (not a directory)
    if os.path.isfile(file_path):
        file_name = os.path.basename(file_path)
        
        code = setup('./went_through_mine_worked/' + file_name)
        if code is None:
            continue

    # Create a new Jupyter Notebook
    notebook = nbformat.v4.new_notebook()
    # Add code cells to the notebook
    x = 0
    for c in code:
        cell = nbformat.v4.new_code_cell(source=c)
        notebook.cells.append(cell)
        x += str(c).count('\n') 

    lines_nb.append(x + len(code))

    with open('./output_notebooks/went_through_mine_worked/' + file_path[27:], 'w', encoding='utf-8') as notebook_file:
        nbformat.write(notebook, './order_dependent_testing/' + file_path[27:])
        print(f"Jupyter Notebook saved to: {file_path}")


```

```python
import numpy as np

print(np.average(total_num_cells))

print(np.average(lines_changed))

print(np.average(lines_nb))
```
