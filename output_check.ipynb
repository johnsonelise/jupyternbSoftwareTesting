#### check the output of both notebooks to see if they are the same

```python
import nbformat
import sys
import matplotlib.pyplot as plt
import random
import glob
import os.path


folder_path = "./final_valid_notebooks//"

pattern = os.path.join(folder_path, '*')
file_number = 0
error = [0, 0]

print(f'total files: {len(glob.glob(pattern))}')
# Iterate through the files
for file_path in glob.glob(pattern):

    file_number = file_number + 1
    # Check if it's a file (not a directory)
    if os.path.isfile(file_path):
        file_name = os.path.basename(file_path)

        with open(folder_path + file_name, 'r', encoding='utf-8') as notebook_file:
            notebook_content_theirs = nbformat.read(notebook_file, as_version=4)
        try:
            with open('./went_through_mine_worked/' + file_name, 'r', encoding='utf-8') as notebook_file:
                notebook_content_mine = nbformat.read(notebook_file, as_version=4)
        except Exception as e:
            error[1] += 1
            print(str(e))
            continue

        for cell, cell_mine in zip(notebook_content_theirs['cells'], notebook_content_mine['cells']):
            if  cell['cell_type'] == 'code' and 'outputs' in cell:
                if cell['outputs'] != cell_mine['outputs']: # figure out a way to put the cell output of other notebook to check
                    error[0] += 1
                    print(f'ERROR: {error[0]}')
                    break

print(f'number of errors:      {error[0]}')
print(f'number of file errors: {error[1]}')
```
