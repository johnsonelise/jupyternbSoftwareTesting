```python
x = linspace(0, 10, 100)
t = linspace(0, 10, 100)
y0 = 1
k = 0.5
ω = 1
```

```python
y = 2*y0*cos(ω*t)*sin(k*x)
```

```python
y
```

```python
f1 = figure(figsize=(8, 8))
a1 = f1.add_subplot(111, projection='3d')

a1.plot(x, t, y);
```
