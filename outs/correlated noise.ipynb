```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cov(x, y):
    xbar = np.sqrt(x**2 + y**2)
    return 2*np.exp( -(xbar/0.05) )

cov(0,0)
```

```python
n = 64
Cs = np.zeros((n**2, n**2))

xs = np.linspace(0.,1.,n)
Xs, Ys = np.meshgrid(xs , xs)
XYs = np.vstack([Xs.ravel(), Ys.ravel()])

for i in xrange(n**2):
    tmp = XYs.T - XYs[:, i]
    res = cov(tmp[:,0], tmp[:,1])
    Cs[:, i] = res

L = np.linalg.cholesky(Cs)
    
plt.imshow(Cs)
```

```python
out = np.dot(L, np.random.normal(size = n**2))
plt.imshow(out.reshape(n,n))
```
