```python
from __future__ import division
%matplotlib inline
import matplotlib as mpl, scipy as sp, numpy as np, pandas as pd, seaborn as sns
from matplotlib import pyplot as plt
```

```python
def plotseries(limit,d,flag):
    # go all the way up to 'limit', 
    # which is limit+1 in a half-open interval like Python uses.
    # flag tells it which graph to plot and where (flag == 1 plots forcing function)
    # First set up the forcing function
    Fo=8
    w1 = 9
    T = 2*np.pi/w1
     
    #set up the graph - in particular, make t axis accomodate several cycles
    t = np.linspace(0,2.1*T,1000)
    y = np.zeros_like(t) # start with y = 0, as an array.
    
    #now set parameters of the system being forced
    m=0.6
    # d is set in the parameters
    for n in range(0,limit+1):
        if flag==1: #If fn 0, don't convert to response; show the forcing function
            Cn = 1
        else: 
            wn = n*w1
            Cn = 1/(d+1j*m*wn)
        if n == 0:                
            y = 1/4*Cn
        else: 
            forceComponent = np.sin(n*np.pi/2)/(np.pi*n)
            forceComponent = forceComponent+(-1j)*(1-np.cos(n*np.pi/2))/(np.pi*n)
            y=y + forceComponent*np.exp(1j*n*w1*t)*Cn
           
    plt.figure(figsize=(10,10))
    plt.subplot(3,1,flag)
    plt.plot(t,Fo*y.real)

terms = 100;
plotseries(terms,1,1)
plotseries(terms,7.2,2)
plotseries(terms,0.5,3)
```
